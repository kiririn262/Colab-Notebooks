# Matplotlibの用語
figure：グラフの全体，キャンパス

Axes：一つのグラフ

X axis（X軸）：横軸

Y axis（Y軸）：縦軸

X axis label：横軸のラベル

Y axis label：縦軸のラベル

X axis ticks：横軸の目盛り

Y axis ticks：縦軸の目盛り

title：グラフのタイトル

legend：グラフの説明

grid：罫線

# 折れ線グラフの作成

```python
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
# グラフのキャンパスとグラフのエリアを作成
fig, ax = plt.subplots(facecolor="white") # 戻り値がfigureオブジェクト，Axes subplotオブジェクト，キャンバスを白くするように引数

# リスト
date = ['月','火','水','木','金','土','日']
temperature = ['19','22','24','21','20','22','20']

ax.plot(date, temperature)
# ax.plot(['月','火','水','木','金','土','日'], ['19','22','24','21','20','22','20'])
```
