# 特定の品詞だけ抽出

```python
import MeCab

t = MeCab.Tagger()
node = t.parseToNode('私は学生です！') # 文章を分割した語の情報

# node.surface：表層文字列
# node.feature：品詞や活用系など
# node.next：次のノード（最後はNone）

while node:
  # 名詞だけをカンマ区切りで表示する
  f = node.feature # 分割した語の情報（カンマ区切り）
  p = f.split(',')[0] # カンマ区切りでリストを作成，品詞の情報を代入
  if p == '名詞':
    print(node.surface) # 名詞なら表層文字列を表示
  node = node.next #最後のノードまで繰り返す
```

```python
import MeCab

t = MeCab.Tagger()
node = t.parseToNode('私はペンを持っています')
while node:
  print(node.feature)
  node = node.next
```

```python
import MeCab

t = MeCab.Tagger()
node = t.parseToNode('私はペンを持っています')
while node:
  f = node.feature # 何番目の値に何が表示されるかはunidic-liteのインストールフォルダにあるdicrcファイルに定義が書いてある．
  p = f.split(',')[0]
  if p == '名詞':
    print(node.surface)
  node = node.next
```
