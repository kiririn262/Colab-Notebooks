#tqdm：進捗確認

```python
# 通常
import time

for i in range(10):
  # 1秒待つ
  time.sleep(1)
```

```python
# tpdmを使う
import tqdm

for i in tqdm.tqdm(range(10)):
  # 1秒待つ
  time.sleep(1)
```
