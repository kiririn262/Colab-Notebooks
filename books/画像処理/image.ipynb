```python
# ライブラリをインポート
import cv2
from matplotlib import pyplot as plt
```

# 1枚の画像を処理

```python
# 画像読み込み
img = cv2.imread('3.mn03.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

## 白黒に変換

```python
# 白黒に変換
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 描画
plt.gray()
plt.imshow(img_gray)
```

## 回転

```python
# 時計回りに90°回転
img_90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_90)
```

## 拡大・縮小

```python
# 画像のサイズを変更する
# 幅
width = 100
# 高さ
height = 100

# 画像のサイズを変更
img_small = cv2.resize(img, (width,height))

# 描画
plt.imshow(img_small)
```

# 一度に複数の画像を処理

```python
# 必要なライブラリをインポート
import glob
import os

# 入力画像のファイルパス一覧
file_path_list = glob.glob('/drive/MyDrive/Colab Notebooks//*')

# ファイルパスを確認
file_path_list
```

```python
# 複数の画像を処理
for file_path in file_path_list:
  # 画像の読み込み
  img = cv2.imread(file_path)

  # ここに処理を記載
  
  img_90= cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

  # ファイル名だけを取得
  file_name = os.path.split(file_path)[1]

  # 保存
  output_path = '/output' + file_name
  cv2.imwrite(output_path, img_90)
```

```python
# 画像読み込み
file = "downloads/図形/8.o1045057014195954149.jpg.png"
img = cv2.imread(file)

# 白黒に変換
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 描画
plt.gray()
plt.imshow(img_gray)
```

```python
# 境界抽出
contours, hierarchy = cv2.findContours(img_gray, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# 描画
cont_img = cv2.drawContours(img, contours , -1, (0,255), 5) # 元画像，抽出した輪郭，-1，最小値・最大値，輪郭の太さ
plt.imshow(cont_img)
```
