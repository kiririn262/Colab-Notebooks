```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
!wget https://upload.wikimedia.org/wikipedia/commons/thumb/4/42/Balaeniceps_rex_2.jpg/250px-Balaeniceps_rex_2.jpg
```

```python
img = cv2.imread("250px-Balaeniceps_rex_2.jpg")

plt.axis('off')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
cv2.imwrite("out_img.jpg", img)
out_img = cv2.imread("out_img.jpg")
plt.axis('off')
plt.imshow(cv2.cvtColor(out_img, cv2.COLOR_BGR2RGB))
```
